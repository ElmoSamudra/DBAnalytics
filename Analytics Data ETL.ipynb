```python
import psycopg2

import pandas as pd

from sqlalchemy import create_engine
```

```python
# Create an engine instance

alchemyEngine   = create_engine('postgresql+psycopg2://shawn:password@127.0.0.1/analytics', pool_recycle=3600);

 

# Connect to PostgreSQL server

dbConnection    = alchemyEngine.connect();
```

```python
# Read data from PostgreSQL database table and load into a DataFrame instance

df = pd.read_sql("select * from event where country = 'Japan'", dbConnection);
```
